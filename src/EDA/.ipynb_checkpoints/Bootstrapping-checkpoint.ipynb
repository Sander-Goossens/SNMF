## Bootstrap per KO

```python
def zou_bootstrap(zou_u, L, zou_SD = '', sample_count = False):
    # L = Number of bootstrapped samples per KO
#     L = 10
    # Number of mutations based on mean total mutation count
    # N = 0
    np.random.seed(0) 
    
    feat = zou_u.reset_index().columns
    genes = zou_u.index
    samples_boot = []

    for i in genes:
        for l in range(L):
            samples_boot.append(i + '_' + str(l))
    zou_boot = pd.DataFrame(0, index = samples_boot, columns = feat)
    
    for index, row in zou_u.iterrows():
        gene = index
        N = round(int(row['Count']))
            
        for l in range(L):
            if sample_count:
                # Sample from Normal distribution --> Poisson or Negative Binomial?
                N = round( int(row['Count']) + np.random.randn()* int(zou_SD.loc[index,'Count']) )  
#             print(round(int(row['Count'])), N, zou_SD.loc[index,'Count'])
#             print(round(int(row['Count'])), np.square(zou_SD.loc[index,'Count']) )
            sample = gene + '_' + str(l)
            zou_boot.loc[sample, 'Gene_KO'] = index
            zou_boot.loc[sample, 'Protein_KO'] = row['Protein_KO']
            zou_boot.loc[sample, 'Subpathway_KO'] = row['Subpathway_KO']
            zou_boot.loc[sample, 'Count'] = N

            Pr = row[feat[3:99]]
            Pr = Pr.tolist()

            muts = np.random.choice(feat[3:99], N, replace = True,  p=Pr)
            for m in muts:
                # Add bootstrapped/sampled mutations to count matrix
                zou_boot.loc[sample, m] += 1
    
    zou_boot[feat[3:99]] = zou_boot[feat[3:99]].div(zou_boot['Count'].values,axis=0)
    
    return zou_boot




zou_boot = zou_bootstrap(zou_u, 10, zou_SD = zou_SD, sample_count = True)

# zou_boot.to_pickle("data/zou2021/zou_boot_{}_{}.pkl".format(L, 'FixedCount'))
# zou_boot = pd.read_pickle("data/zou2021/zou_boot_10_FixedCount.pkl")

zou_boot


            

            
            
```

```python
def SBS_feat():
    zou = pd.read_pickle("data/zou2021/zou_SBS_indel_DBS.pkl")
    features = zou.columns[0:96]
    return features

def indel_feat():
    indel_features = []
    for i in range(1, 84):
        indel_features.append(str(i))
    return indel_features

def DBS_feat():
    xls = pd.ExcelFile('data/zou2021/DBS_features.xlsx')
    DBS_feat = pd.read_excel(xls, 'Blad1', index_col= 0,header= None)
    DBS_features = DBS_feat.index.tolist()
    return DBS_features

def screeplot(pca,title):
    var = pca.explained_variance_ratio_
    cumsum = np.cumsum(var)
    plt.plot(np.arange(pca.n_components)+1, var, 'ro-')
    plt.plot(np.arange(var.size) + 1, cumsum)
    plt.title('Scree Plot')
    plt.xlabel('PC')
    plt.ylabel('Proportion of Variance explained')
    plt.show()
#     plt.savefig('plots/multi/_{0}_screeplot_{}.png'.format(pca.n_components))

def EDA_PCA(zou, n_comp, title):
    
    #
    if title[1] == 'SBS':
        features = SBS_feat()
    elif title[1] == 'Indel':
        features = indel_feat()
    elif title[1] == 'DBS':
        features = DBS_feat
    elif title[1] == 'SBS+Indel':
#         SBS_feat().type
#         indel_feat().type
        features = [*SBS_feat(), *indel_feat()]
        
    
    x = zou.loc[:,features].values  # Separating out the target
    x = StandardScaler().fit_transform(x)
    
    pca = PCA(n_components=n_comp)
    pc = pca.fit_transform(x)
    principal_df = pd.DataFrame(data= pc, columns= np.arange(n_comp)+1)

    if title[0] == 'Zou_COMB':
        final_df = pd.concat([principal_df, zou[['Gene_KO','Subpathway_KO','Count', 'Gene_KO_paired','Subpathway_KO_paired']].reset_index(drop=True)], axis =1)
        pal = 'Paired'
    else:
        final_df = pd.concat([principal_df, zou[['Gene_KO','Subpathway_KO','Count']].reset_index(drop=True)], axis =1)
        pal = 'bright'
        
    plt.figure(figsize=(12,8))

    num_labels = len(set(zou.loc[:, [title[2]]].values.flatten()))
    
    sns.scatterplot(
        x= 1 , y= 2,
        hue=title[2],
        size = "Count",
        sizes=(50,500),
        palette=sns.color_palette(pal, num_labels),
        data=final_df,
        alpha=0.9,
        legend ='auto'
    )
    
    plt.title('PCA - {0} - {1} - {2} - {3}'.format(*title))
    plt.xlabel('PC 1')
    plt.ylabel('PC 2')
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()

    screeplot(pca, title)  
#     print(final_df)
    
    
def EDA_UMAP(zou, n_comp, title, metric='euclidean', n_neighbors= 15):
    '''
    UMAP settings:
    n_neighbors : 15 (default)
        local (low value) vs global (high value) structure
    min_dist : 0.1 (default)
        minimum distance between points in low dim
    n_components : 2
        number of dimensions
    metric : 'cosine'
    '''
    if title[1] == 'SBS':
        features = SBS_feat()
    elif title[1] == 'Indel':
        features = indel_feat()
    elif title[1] == 'DBS':
        features = DBS_feat
    elif title[1] == 'SBS+Indel':
#         SBS_feat().type
#         indel_feat().type
        features = [*SBS_feat(), *indel_feat()]
        
    x = zou.loc[:,features].values  # Separating out the target
    x = StandardScaler().fit_transform(x)

    reducer = umap.UMAP(random_state=42, n_components=n_comp, metric=metric, n_neighbors = n_neighbors)
    embedding = reducer.fit_transform(x)

    final_df = pd.DataFrame(data=embedding, columns=np.arange(n_comp) + 1)
    if title[0] == 'Zou_COMB':
        final_df = pd.concat([final_df, zou[['Gene_KO','Subpathway_KO','Count', 'Gene_KO_paired','Subpathway_KO_paired']].reset_index(drop=True)], axis =1)
        pal = 'Paired'
    else:
        final_df = pd.concat([final_df, zou[['Gene_KO','Subpathway_KO','Count']].reset_index(drop=True)], axis =1)
        pal = 'bright'
    
    
    plt.figure(figsize=(12,8))
    
    num_labels = len(set(zou.loc[:, [title[2]]].values.flatten()))
    
    sns.scatterplot(
        x= 1 , y= 2,
        hue=title[2],
        size = "Count",
        sizes=(50,500),
        palette=sns.color_palette(pal, num_labels),
        data=final_df,
        alpha=0.9,
        legend ='auto'
    )
    
    plt.title('UMAP - {0} - {1} - {2} - {3}'.format(*title))
    plt.xlabel('axis 1')
    plt.ylabel('axis 2')
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    plt.show()
```

## Bootstrap per SAMPLE

```python
def zou_bootstrap_perSample(zou, L, zou_SD = '', sample_count = False):
    # L = Number of bootstrapped samples per KO
#     L = 10
    # Number of mutations based on mean total mutation count
    # N = 0
    np.random.seed(0) 
    
    feat = zou.reset_index().columns
    genes = zou.index
    samples_boot = []

    for i in genes:
        for l in range(L):
            samples_boot.append(i + '_' + str(l))
    zou_boot = pd.DataFrame(0, index = samples_boot, columns = feat)
    
    for index, row in zou.iterrows():
        gene = index
        N = round(int(row['Count']))
            
        for l in range(L):
            if sample_count:
                # Sample from Normal distribution --> Poisson or Negative Binomial?
                N = round( int(row['Count']) + np.random.randn()* int(zou_SD.loc[index,'Count']) )  

            sample = gene + '_' + str(l)
            zou_boot.loc[sample, 'Gene_KO'] = row['Gene_KO']
            zou_boot.loc[sample, 'Protein_KO'] = row['Protein_KO']
            zou_boot.loc[sample, 'Subpathway_KO'] = row['Subpathway_KO']
            zou_boot.loc[sample, 'Count'] = N

            Pr = row[feat[1:97]]
            Pr = Pr.tolist()

            muts = np.random.choice(feat[1:97], N, replace = True,  p=Pr)
            for m in muts:
                # Add bootstrapped/sampled mutations to count matrix
                zou_boot.loc[sample, m] += 1
    
    zou_boot[feat[1:97]] = zou_boot[feat[1:97]].div(zou_boot['Count'].values,axis=0)
    
    return zou_boot


L = 5

zou = pd.read_pickle("data/zou2021/zou_96SBS_filtered.pkl")

zou_boot = zou_bootstrap_perSample(zou, L, zou_SD = '', sample_count = False)

zou_comb = pd.concat([zou_boot, zou])


for index, row in zou_comb.iterrows():
    if len(index.split('_')) == 4:
        zou_comb.loc[index, 'Gene_KO_paired'] = row['Gene_KO'] + '_' + 'boot'
        zou_comb.loc[index, 'Subpathway_KO_paired'] = row['Subpathway_KO'] + '_' + 'boot'
    else:
        zou_comb.loc[index, 'Gene_KO_paired'] = row['Gene_KO'] + '_' + 'real'
        zou_comb.loc[index, 'Subpathway_KO_paired'] = row['Subpathway_KO'] + '_' + 'real'


zou_comb = zou_comb.sort_values(by=['Subpathway_KO_paired'])
zou_comb
```
